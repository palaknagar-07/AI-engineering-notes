# Zero-Centered Data & Why It Helps (Quick Interview Notes)

## Core Idea
Zero-centered data (mean ≈ 0) makes gradient updates **balanced and unbiased**, leading to **faster and more stable learning**.

---

## Why it matters in backprop
Gradient for a weight:
\[
\frac{dL}{dw} = \frac{dL}{dz} \cdot x
\]

So input `x` directly controls gradient direction.

---

## Problem (non-zero-centered data)
- If `x > 0` always (e.g., sigmoid output 0–1):
  - Gradients mostly have same sign
  - Updates push weights in a biased direction
  - Optimization zig-zags and is slower

---

## Solution (zero-centered data)
- If `x` has positive + negative values:
  - Gradients cancel out partially
  - Updates are balanced
  - Faster convergence and smoother optimization

---

## Why sigmoid is worse than tanh
- Sigmoid: outputs (0, 1) → not zero-centered
- Tanh: outputs (-1, 1) → zero-centered → better gradient flow

---

## Why we standardize inputs
\[
x' = \frac{x - \mu}{\sigma}
\]

### Benefits:
- Mean becomes 0 → balanced gradients
- Same scale for all features → stable optimization
- Faster convergence in gradient descent

---

## One-line intuition
> Zero-centered inputs prevent biased gradient flow, making weight updates more symmetric and optimization faster.

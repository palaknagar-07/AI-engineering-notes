# Useful APIs for AI, RAG & Agent Engineering

|------|-------------|------------|------------|

| Firecrawl | Convert websites into clean Markdown | Free tier + Paid | Website RAG, web scraping for LLMs |
| Unstructured | Parse PDFs, DOCX, PPT, emails | Free tier + Paid | Enterprise document ingestion |
| LlamaParse | Extract text/tables from complex PDFs | Free tier + Paid | Research papers, invoices, reports |
| Tavily | AI-native web search | Free tier + Paid | AI agents, RAG search |
| Exa | Semantic search engine | Free tier + Paid | Research agents, knowledge retrieval |
| SerpAPI | Google, Maps, YouTube search APIs | Free tier + Paid | Search-based applications |
| AssemblyAI | Speech-to-text and audio intelligence | Free tier + Paid | Meeting notes, voice assistants |
| Deepgram | Fast speech recognition | Free tier + Paid | Real-time voice applications |
| Google Maps API | Maps, directions, places | Free tier + Paid | Location-based apps |
| OpenStreetMap APIs | Open-source map data | Free | Mapping applications |
| Gmail API | Read/send/manage emails | Free | Email agents and automation |
| Google Drive API | Access and manage Drive files | Free | Document management agents |
| Google Calendar API | Read/create calendar events | Free | Scheduling assistants |
| Resend | Email sending service | Free tier + Paid | Notifications and transactional emails |
| NewsAPI | News articles and headlines | Free tier + Paid | News aggregation and RAG |
| Alpha Vantage | Stock and forex data | Free tier + Paid | Finance dashboards |
| Polygon.io | Professional market data | Paid (limited free) | Fintech applications |
| Hugging Face Inference API | Run ML models via API | Free tier + Paid | AI prototypes and ML apps |
| Cohere API | Embeddings and reranking | Free tier + Paid | RAG retrieval improvement |
| Jina AI Reader | Convert webpages to clean text | Free | Web content extraction for RAG |
| OpenAI API | LLMs, embeddings, vision | Paid | Chatbots, agents, AI products |
| Anthropic API | Claude models | Paid | Long-context reasoning systems |
| Groq API | Ultra-fast LLM inference | Free tier + Paid | Low-latency AI applications |
| Together AI | Hosted open-source models | Free tier + Paid | Cost-effective AI deployment |
| Replicate | Run open-source AI models | Pay-as-you-go | Image, video, audio AI |
| Pinecone | Managed vector database | Free tier + Paid | Production RAG systems |
| Weaviate | Vector database | Free tier + Paid | Semantic search and RAG |
